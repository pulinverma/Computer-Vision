```python
def rgb2gray(rgb):                                    #converting into gray scale
    r,g,b=rgb[:,:,0],rgb[:,:,1],rgb[:,:,2]
    gray=0.2989*r+0.5870*g+0.1140*b
    return gray
```

```python
block1=np.vstack((arr1,arr2,arr3,arr4,arr5,arr6,arr7,arr8,arr9,arr10))
print(block1.shape)
block2=np.vstack((arr11,arr12,arr13,arr14,arr15,arr16,arr17,arr18,arr19,arr20))
print(block2.shape)
```

```python
X=np.vstack((block1,block2))
print(X.shape)
y=np.vstack((np.zeros((10,1)),np.ones((10,1))))
print(y.shape)
```

```python
pca=PCA()                               #feature reduction,i.e pca
pca.fit_transform(X)
X.shape
```

```python
X_train1=block1[:6,:]                                  #train and test for fdr
X_train2=block2[:6,:]
X_test1=block1[6:,:]
X_test2=block2[6:,:]
X_train=np.vstack((X_train1,X_train2))
y_train=np.vstack((np.zeros((6,1)),np.ones((6,1))))
X_test=np.vstack((X_test1,X_test2))
y_test=np.vstack((np.zeros((4,1)),np.ones((4,1))))
fdr=[]                                                 #feature selection,i.e fdr
s=X_train.shape
for j in range(s[1]):
    x1=X_train[:6,j]
    x2=X_train[6:,j]
    m1=np.mean(x1)
    m2=np.mean(x2)
    s1=np.std(x1)
    s2=np.std(x2)
    if((s1==0)and(s2==0)):
        fdr1=0
    else:
        fdr1=((m1-m2)**2)/((s1**2)+(s2**2))
    fdr.append(fdr1)
#print(fdr)        #provide fdr of all features or columns.
fdr=np.array(fdr)   
fdr=-1*fdr         #multiply by -1 for making decreasing order.
#print(fdr)
q=np.argsort(fdr)  #provide indices of fdr with dec. order (actually argsort is used for inc. order).
#print(q)
```

```python
X_train=X_train[:,q]                                    #after feature selection train and test
X_train=X_train[:,:300]
X_test=X_test[:,q]
X_test=X_test[:,:300]
```

```python
arr=np.random.permutation(12)                           #shuffling
X_train=X_train[arr,:]
y_train=y_train[arr]
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
acc_set=[]
spec_set=[]
sens_set=[]
for feat_set in range(300): 
    X_T1=X_train[:,:(feat_set+1)]
    clf=MLPClassifier(solver='lbfgs',alpha=1e-3,hidden_layer_sizes=(80,2),random_state=1)
    clf.fit(X_T1,y_train)
    predicted=clf.predict(X_test[:,:(feat_set+1)])
    TP=0
    TN=0
    FN=0
    FP=0
    for i in range(len(y_test)):
        if(y_test[i]==predicted[i]):
            if(y_test[i]==1):
                TP+=1
            else:
                TN+=1
        else:
            if(predicted[i]==1):
                FP+=1
            else:
                FN+=1
    acc=(TP+TN)/(TP+TN+FP+FN)      #accuracy
    acc_set.append(acc)
    sens=TP/(TP+FN)                #sensitivity
    sens_set.append(sens)
    spec=TN/(TN+FP)                #specificity
    spec_set.append(spec)
```

```python
print(np.max(acc_set),'   ',np.max(sens_set),'   ',np.max(spec_set))
```

```python
plt.plot(acc_set,color='red',marker='o',markersize='1',markeredgecolor='green',linewidth=1)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
plt.plot(sens_set,color='purple',marker='s',markersize='1',markeredgecolor='blue',linewidth=1)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
plt.plot(spec_set,color='orange',marker='^',markersize='1',markeredgecolor='black',linewidth=1)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```
